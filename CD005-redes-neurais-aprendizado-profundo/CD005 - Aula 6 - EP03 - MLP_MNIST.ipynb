Multi-layer perceptron no Dataset MNIST

Primeiro, importe o tensorflow, carregue e normalize o dataset

```
import tensorflow as tf
import time

mnist = tf.keras.datasets.mnist

(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255, x_test / 255
```

Criação e preparação de um MLP com uma camada oculta

```
simple_model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),       # 1 camada de entrada (achatando a entrada pra ela virar um vetor)
    tf.keras.layers.Dense(3, activation=tf.nn.relu),    # 1 camada oculta com 3 neuronios
    tf.keras.layers.Dense(10, activation=tf.nn.softmax) # 1 camada oculta com 10 saidas (porque?)
])


simple_model.compile(optimizer='rmsprop',               # RMSProp de otimizador
              loss='sparse_categorical_crossentropy',   # entropia cruzada adaptada para multiplas classes
              metrics=['accuracy'])                     # métrica final de desempenho

simple_model.summary()
```

Função que recebe um modelo, os dados, o treina por um numero de epocas, o avalia e diz o tempo gasto no treino

```
def train_and_test(model, x_train, y_train, x_test, y_test, epocas=10): 
  # aqui vc vai dar start em um 'cronometro' pra medir o tempo gasto
  start = time.time()  # troque o 0 pelo comando que faz a tomada de tempo
  
  model.fit(x_train, y_train, epochs=epocas) # faz o treino do modelo

  # aqui vc vai paralizar o 'cronometro' e exibir o tempo gasto
  duration = time.time() - start  # troque o 0 pelo comando que faz a tomada de tempo
  
  print(f"Treinamento demorou {duration} segundos")
  
  print(f"Avaliação do modelo:")
  model.evaluate(x_test, y_test)

```

Vamos chamar a função para o modelo simples instanciado anteriormente

```
train_and_test(simple_model, x_train, y_train, x_test, y_test, 10)
```
