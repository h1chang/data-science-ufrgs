Neste colab você irá implementar um Neurônio Artificial. Inicialmente, o peso e o limiar são definidos internamente. Depois, eles serão expostos como parâmetros para facilitar que você "brinque" com os valores

```python
# Execute esta celula para carregar as funções, mas não se preocupe com o conteudo
# São apenas funções auxiliares para os testes automatizados

def testa_float(esperado:float, obtido:float, tol=1E-9) -> None:
  """
  Compara se a diferenca entre 2 floats esta dentro da tolerancia e imprime
  a mensagem correspondente
  """
  if abs(esperado - obtido) < tol:
    print("Passou no teste")
  else:
    print(f"Erro. Esperado: {esperado}, obtido: {obtido}")


def testa_int(esperado:int, obtido:int):
  if esperado == obtido:
    print("Passou no teste")
  else:
    print(f"Erro. Esperado: {esperado}, obtido: {obtido}")
```

Implemente abaixo o neurônio limiar, conforme visto em aula. Teste-o na célula seguinte.

```python
def neuronio_limiar(x: float) -> int:
  """
  Implementa um neuronio limiar com peso e limiares internos 
  (mais proximo ao neuronio biologico)
  :param x: sinal de entrada do neuronio
  :return: sinal de saida do neuronio
  """
  w = 1
  limiar = 2

  # inicio: implemente o codigo que faça o neuronio retornar 1 se w*x >= limiar, e 0 caso contrario
  if w*x >= limiar:
    return 1
  else:
    return 0
  # fim


# --- a chamada abaixo executa a funcao com a entrada especificada e mostra a saida
print(neuronio_limiar(1))
print(neuronio_limiar(2))
print(neuronio_limiar(3))

```

```python
def testa_neuronio_limiar():
  # para executar o teste, coloque w = 1 e limiar = 2 dentro do seu neuronio_limiar
  testa_int(0, neuronio_limiar(1))
  testa_int(1, neuronio_limiar(3))

#### --- a chamada abaixo executa o teste. a msg "Passou no teste" deve aparecer 2 vezes---
testa_neuronio_limiar()
```

Biologicamente, os pesos e o limiar fazem parte do neurônio. Porém, em termos de programação, é mais fácil expor os pesos e o limiar para serem definidos fora da função, para facilitar o processo de ajuste dos pesos. Implemente o neurônio abaixo e teste-o na célula seguinte.

```python
def neuronio_limiar_parametrizado(w1:float, w2:float, limiar:float, x1: float, x2: float) -> int:
  """
  Implementa um neuronio limiar com 2 entradas, 2 pesos e 1 limiar
  :param w1: peso aplicado ao 1o sinal de entrada 
  :param w2: peso aplicado ao 2o sinal de entrada 
  :param limiar: limiar de ativação do neuronio
  :param x1: 1o sinal de entrada do neuronio
  :param x2: 2o sinal de entrada do neuronio
  :return: sinal de saida do neuronio
  """
  # inicio: implemente o codigo que faça o neuronio retornar 1 se pesos*entradas >= limiar, e 0 caso contrario
  if (w1*x1 + w2*x2 >= limiar):
    return 1
  else:
    return 0
  # fim


# --- a chamada abaixo executa a funcao com a entrada especificada e mostra a saida
print(neuronio_limiar_parametrizado(w1=1, w2=2, limiar=2, x1=1, x2=1))
```

```python
# NAO MUDE O CONTEUDO DESSA CELULA, apenas execute-a para testar a celula anterior
def testa_neuronio_limiar_parametrizado():
  # teste 1 -- neuronio nao ativa
  testa_int(0, neuronio_limiar_parametrizado(w1=0, w2=0, limiar=1, x1=1, x2=1000))
  # teste 2 -- neuronio ativa
  testa_int(1, neuronio_limiar_parametrizado(w1=1.5, w2=0.5, limiar=1, x1=2, x2=4))

  
#### --- a chamada abaixo executa o teste
testa_neuronio_limiar_parametrizado()
```

**Muitas entradas**

Até então, os neurônios implementados tem dois parâmetros para cada entrada (o valor dela e o peso correspondente). Mas como fazer para lidar com muitas entradas? Na próxima célula, implemente um neurônio artificial que computa sua ativação para qualquer número de entradas.

Observe que agora os pesos e as entradas (x) são arrays especificados como arrays numpy, permitindo que o neurônio receba um número de entradas arbitrário, desde que estejam em um array.

```python
import numpy as np 

def neuronio_limiar_multidimensional(w:np.array, limiar:float, x:np.array) -> float:
  """
  Implementa um neuronio limiar com 2 entradas, 2 pesos e 1 limiar
  :param w: vetor de pesos
  :param limiar: limiar de ativação do neuronio
  :param x: vetor com os sinais de entrada
  :return: sinal de saida do neuronio
  """
  # inicio: implemente o codigo que faça o neuronio retornar 1 se pesos*entradas >= limiar, e 0 caso contrario
  tamanho = len(w)
  somatorio = 0
  for i in range(tamanho):
    # print(i)
    somatorio = somatorio + (w[i] * x[i])

  if (somatorio >= limiar):
    return 1
  else:
    return 0
  
  
  # fim


# --- a chamada abaixo executa a funcao com a entrada especificada e mostra a saida
# note que as entradas correspondem aos valores do segundo teste do neuronio_limiar(w1, w2, limiar, x1, x2) 
# você pode inventar novos w e x, desde que tenham o mesmo numero de elementos
neuronio_limiar_multidimensional(w=np.array([1.5, 0.5]), limiar=1, x=np.array([2, 4]))
```

```python
# NAO MUDE O CONTEUDO DESSA CELULA, apenas execute-a para testar a celula anterior
def testa_neuronio_limiar_multidimensional():
  # teste 1 -- neuronio nao ativa
  testa_int(0, neuronio_limiar_multidimensional(w=np.array([0, 0]), limiar=1, x=np.array([1,1000]) ) )
  # teste 2 -- neuronio ativa
  testa_int(1, neuronio_limiar_multidimensional(w=np.array([1.5, 0.5]), limiar=1, x=np.array([2, 4]) ) )

  
#### --- a chamada abaixo executa o teste
testa_neuronio_limiar_multidimensional()
```
