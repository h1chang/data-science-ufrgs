## Usando máscaras para alterar os dados

### #5
Para o padrão abaixo, aplique uma máscara para alterar seus valores para -1.

```python
url='https://www.inf.ufrgs.br/~schnorr/cd009/ad04_5.png'
Disp.Image(requests.get(url).content)
```

```python
# 1. Calcular as coordenadas na forma de uma máscara 2D
# 2. Alterar os valores a partir das máscara criada
a = np.arange(100).reshape(10, 10)

v = np.arange(1, a.shape[0] -1, 2)
u = v + 1
mask = [v, u]
a[mask]
```
