# Operações comuns

Vamos ver algumas operações comuns.

Selecionar os valores únicos da coluna `Year`.

```python
df["Year"].unique()
```

Selectionar algumas colunas.

```python
df[["Origin", "Dest", "DepDelay"]]
```

Como remover algumas colunas

```python
df.drop(columns=["Year", "Month", "DayofMonth"])
```

Vamos alterar uma coluna:

```python
df["DayOfWeek_v0"] = df["DayOfWeek"] - 1
df
```

Podemos filtrar as linhas baseado em condições

```python
df[(df["DayofMonth"] == 1) & (df["Month"] == 12)]
```

Podemos fatiar linhas com, por exemplo para pegar as linhas 532, 533, 534 e 535 dos dados:

```python
df[532:536] # Como em NumPy, o limite superior não é incluído
```

Podemos reordenar os dados por colunas.

```python
df.sort_values(["Month", "DayofMonth"])
```

Ou aindar ordenar de maneira decrescente, por exemplo para mostrar qual foi o vôo de 1987 com o maior atraso (SFO até SNA, com 1033 minutos de atraso na chegada):

```python
df[["ArrDelay", "Origin", "Dest"]].sort_values(["ArrDelay"], ascending=False)
```

Podemos criar grupos com `groupby`. Por exemplo, vamos agrupar os dados por `Origin` e `Dest` para percebermos que existem 3464 pares origem/destino de vôos em 1987.

```python
df.groupby(["Origin", "Dest"]).size()
```

Podemos empregar `groupby` e obter qual foi o maior atraso para cada uma das rotas Origem/Destino utilizando a função `agg`. Para terminar, ordenamos pelo pior atraso para saber qual o par Origem/Destino que tem o maior atraso.

```python
df.groupby(["Origin", "Dest"]) \
  .agg({"ArrDelay": 'max'}) \
  .sort_values(["ArrDelay"], ascending=False)
```

O uso do encadeamento com o operador `.` e a contrabarra `\` permite calcular de maneira relativamente legível um sumário estatístico dos atrasos, em conjunto com a operação `describe`.

```python
df.groupby(["Origin", "Dest"]) \
  .agg({"ArrDelay": 'max'}) \
  .sort_values(["ArrDelay"], ascending=False) \
  .describe()
```

## Simples

Qual a empresa que tem mais vôos não cancelados?

```python
df[(df["Cancelled"] == 0)].groupby("UniqueCarrier").size()
```

Qual a `UniqueCarrier` que tem o maior atraso de chegada (`ArrDelay`)?

```python
df[(df["Cancelled"] == 0)] \
     .groupby("UniqueCarrier") \
     .ArrDelay \
     .max() \
     .sort_values(ascending=False) \
     .iloc[0]
```

Qual o dia da semana que tem mais vôos?

```python
df[(df["Cancelled"] == 0)] \
    .groupby("DayOfWeek") \
    .size() \
    .sort_values(ascending=False) \
    .index[0]
```

Qual a empresa que apresentar o maior número de cancelamentos?

```python
df[(df["Cancelled"] == 1)] \
    .groupby("UniqueCarrier") \
    .size() \
    .sort_values(ascending=False) \
    .index[0]
```

Qual empresa que voou mais milhas?

```python
df[(df["Cancelled"] == 1)] \
    .groupby("UniqueCarrier") \
    .Distance \
    .sum() \
    .sort_values(ascending=False) \
    .index[0]
```

## Avançados

Qual a empresa `UniqueCarrier` que conseguiu recuperar melhor os atrasos durante o vôo?

```python
dft = df[(df["Cancelled"] == 0)][["UniqueCarrier", "DepDelay", "ArrDelay"]]
dft["Recover"] = dft["DepDelay"] - dft["ArrDelay"]
dft.groupby("UniqueCarrier")["Recover"] \
    .sum() \
    .sort_values(ascending=False)
```

Há um crescimento no número de vôos?

```python
df[(df["Cancelled"] == 0)] \
    .groupby("Year") \
    .size() \
    .sort_values(ascending=False)
```

```python
dft = df[(df["Cancelled"] == 0)][["UniqueCarrier", "Year", "Distance"]]
dft = dft.groupby(["UniqueCarrier", "Year"]) \
    .size()
# Continuar
```
